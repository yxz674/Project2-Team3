#1 read cases deaths by State RDF file

```python
from rdfpandas.graph import to_dataframe
import pandas as pd
import rdflib
```

```python
g = rdflib.Graph()
g.parse('Resources\cases_deaths_by_State.rdf')
for index,(sub,pred,obj) in enumerate(g):
    print(sub,pred,obj)
    if index == 10:
        break
```

```python
print(f'graph has {len(g)} facts')
```

```python
df = to_dataframe(g)
df.to_csv('cases_deaths_by_State.csv', index = True, index_label = "@id")
```

```python
state_data_path = r'cases_deaths_by_State.csv'
state_data_df = pd.read_csv(state_data_path)
state_data_df.head()
```

```python
organized_df = state_data_df[["ds:state{Literal}","ds:submission_date{Literal}","ds:conf_cases{Literal}","ds:conf_death{Literal}"]]
organized_df.head()
```

```python
renamed_df = organized_df.rename(columns={"ds:state{Literal}":"state", "ds:submission_date{Literal}":"Date","ds:conf_cases{Literal}":"Confirmed_Cases","ds:conf_death{Literal}":"Confirmed_Death"})
renamed_df.head()
```

```python
renamed_df['Month'] = pd.DatetimeIndex(renamed_df['Date']).month
```

```python
renamed_df['Year'] = pd.DatetimeIndex(renamed_df['Date']).year
```

```python
renamed_df['Month']
renamed_df['Year']

clean_df = renamed_df.drop(columns = ['Date']).fillna(0)
clean_df
```

#2 read us_counties_covid19_daily csv file

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
covid_daily_path = r'Resources\us_counties_covid19_daily.csv'
covid_daily_data_df = pd.read_csv(covid_daily_path)
```

```python
covid_daily_data_df.info()
```

```python
covid_daily_data_df.describe()
```

```python
covid_daily_data_df.head()
```

```python
covid_daily_data_df['Month'] = pd.DatetimeIndex(covid_daily_data_df['date']).month
```

```python
covid_daily_data_df['Month']

covid_daily_data_df.head()
```

```python
df_pivot = covid_daily_data_df.pivot_table(index=['state','county'],columns=['Month'],values=['cases','deaths'],aggfunc='count')
cleaned_df = df_pivot.head(10).fillna(0).style.set_precision(0)
cleaned_df
```

```python
cleaned_df.to_csv("cleaned_df.csv", index= False)
```
